```python
import random
import time

headers = []
punks=[]
values = {}
#df = pd.DataFrame()
n=1

data = driver.find_element_by_class_name('table-responsive')

for element in data.find_elements_by_tag_name('th'):
        title = element.text
        headers.append(title)
headers.append("Punk")

def scrap_punks(i):
    
    url= 'https://www.larvalabs.com/cryptopunks/details/'+str(i)
    driver.get(url)
    data = driver.find_element_by_class_name('table-responsive')

    punks_data=data.find_elements_by_tag_name('td')
    punks = [element.text for element in punks_data]
    x = int(len(punks)/5)
    
    for j in range(5):
        values[headers[j]]=punks[j::5]
        values["Punk"] = [i]*x
        
    return pd.DataFrame(values)    

df = scrap_punks(0)

while n<1000:
    df = df.append(scrap_punks(n))
    n = n+1
    time.sleep(random.randint(1,3)) #Script will be stopped for a period of time ranging from 5 to 16 seconds.
```
